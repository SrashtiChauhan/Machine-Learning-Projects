```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, classification_report
```

```python
dataset = pd.read_csv(r"/content/creditcard.csv")
#print(dataset.head(10))

dataset.drop(columns = ['Time'], axis=1, inplace=True)
print(dataset.info())
```

```python
print(dataset['Class'].value_counts())
```

fraud=1   legit=0

```python
print(dataset.describe())
```

```python
legit=dataset[dataset.Class==0]
fraud=dataset[dataset.Class==1]

#print(legit.head(10))
print(legit.value_counts())
print(fraud.value_counts())
```

```python
new_df=legit.sample(n=492)
print(new_df.head(10))
```

```python
new_dataset=pd.concat((new_df,fraud), axis=0)
#print(new_dataset.head(10))
print(new_dataset.shape)
```

```python
x= new_dataset.drop(columns=['Class'], axis=1)
y= new_dataset['Class']
```

```python
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.2,random_state=1)
print(x_train.head(5))
print(y_train.head(5))
```

```python
model= LogisticRegression(max_iter=2000)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_train)
print(y_pred)
print(type(y_pred))
```

```python
accuracy = accuracy_score(y_pred, y_train)
print(f"model accuracy={accuracy*100:2f}%")
```

```python
y_pred =model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"model accuracy={accuracy*100:2f}%")

print(classification_report(y_test, y_pred))
```
